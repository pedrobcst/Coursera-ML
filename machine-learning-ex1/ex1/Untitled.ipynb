```python
def gradientDescent(X, y, theta, iterr, alpha):
    const=alpha/y.size
    cost = []
    for i in range(iterr):
        theta_temp = theta.copy()
        theta_temp = theta_temp - const*(np.transpose(X)*(X*theta - y))
        theta = theta_temp
        cost.append(costFunction(X,y,theta))
    return (cost, theta)
        
```
